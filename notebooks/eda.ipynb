Here is a quick count of the missing data values for each variable. As you can see, there are no missing values, making this a very easy dataset to work with.

```python
def detect_outlier(data):
    cols_list = data.columns.tolist()
    for i in cols_list:
        data_1 = data[i].tolist()
        threshold=3
        mean_1 = np.mean(data_1)
        std_1 =np.std(data_1)
        outliers = []
        for y in data_1:
            z_score= (y - mean_1)/std_1 
            if np.abs(z_score) > threshold:
                outliers.append(y)
        return outliers


       
```

```python
detect_outlier(raw_cc_data)
```

I created a short function to detect the outliers in each variable. The function returns an empty list, meaning there are no outliers in this dataset, making the data set even easier to work with.

In conclusion, this Kaggle dataset is very clean and will not require much transformation.

```python
X = raw_cc_data.iloc[:,0:30]
print(X)
y = raw_cc_data.iloc[:,-1]
```

```python
best_feat = SelectKBest(score_func = f_classif, k=10)
fit = best_feat.fit(X,y)

dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
print(featureScores.nlargest(10,'Score'))  #print 10 best features
```
